### Write a program that will calculate the cost of a custom cup of coffee at a gourmet coffee shop, based on the size of the cup, the type of coffee selected, and flavors that can be added to the coffee. It should complete the following steps:

Ask the user what size cup they want, choosing between small, medium, and large.
Ask the user what kind of coffee they want, choosing between brewed, espresso, and cold brew.
Ask the user what flavoring they want, if any. Choices include hazelnut, vanilla, and caramel.
Calculate the price of the cup using the following values:

Size:
small: \\$2
medium: \\$3
large: \\$4

Type:
brewed: no additional cost
espresso: 50 cents
cold brew: \\$1

Flavoring:
None: no additional cost
All other options: 50 cents


Display a statement that summarizes what the user ordered.
Display the total cost of the cup of coffee as well as the cost with a 15% tip, in phrases that explain the values to the user. Round the cost with tip to two decimal places.
For example, if the user asks for a medium-sized espresso with hazelnut flavoring, the total should be \\$4; the total with a tip should be \\$4.60.


```python
# Collect user input
print("welcome to the BIG coffee shop! We sell many goods: coffee, baked goods and mugs!")
curr_prod = input("what do you want to start first? (coffee/baked goods/mugs)? People usually start with our coffee ")
coffee_cost = 0
if curr_prod != 'coffee':
    print("nope, you are starting with our coffee")
size = input("What size cup would you like? (Small, Medium, Large) ").lower()
coffee_type = input("What kind of coffee would you like? (Brewed, Espresso, Cold Brew) ").lower()
if_flavour = input("Would you like flavoring? (yes/no) ").lower()
if if_flavour == 'yes':
    flavor = input("What would you like for flavoring? (Hazelnut, Vanilla, Caramel) ").lower()

# Calculate base price
if size == "small":
    coffee_cost += 2
elif size == "medium":
    coffee_cost += 3
else:
    coffee_cost += 4

# Add additional cost based on coffee type and flavoring
if coffee_type == "espresso":
    coffee_cost += 0.5
elif coffee_type == "cold brew":
    coffee_cost += 1
if flavor != "no":
    coffee_cost += 0.5

# Calculate total price with 15% tip
coffee_tip_cost = round(coffee_cost * 1.15, 2)
print(coffee_tip_cost)

baked_cost = 0
mugs_cost = 0
curr_prod = input("what do you want next? (baked goods/mugs/none)? You can also walk away now because I only get extra commision from our coffee so I don't care about the other products ")
if curr_prod == 'baked goods':
    print("Uhhhh fine. We only have banana bread. It tastes like garbage and super pricy -- $20 per piece ")
    if_baked_goods = input("you still want it? (yes/no)")
    if if_baked_goods == "yes":
        baked_amt = int(input("You just don't listen, do you? Fine. How many does our BIGSHOT CUSTOMER want? "))
        if baked_amt >= 10:
            print("OH WOW. That's almost how much I can make every day! I'll wrap it up for you")
        baked_cost += baked_amt * 20
        print("now you ordered", baked_amt, "banana breads with $20 each, which is", baked_cost, "in total")
elif curr_prod == 'mugs':
    print("I should warn you, our mug's price is 'mugging' you")
    if_mugs = input("you still want it? (yes/no) ")
    if if_mugs == "yes":
        mugs_amt = int(input("You choose to waste your money... *whispering: and my time*.. They'are $100 each. How many do you want? "))
        if mugs_amt < 0:
            print("hee-heee, GOTCHA! Someone is trying to find a loophole in here... You have now being punished for dishonesty and charged a $10000 fine.")
            mugs_cost = 10000
            print("now you have $10000 charge")
        else:
            mugs_cost += mugs_amt * 100
            print("now you ordered", mugs_amt, "mugs with $100 each, which is", mugs_cost, "in total")
        
else:
    print("Nice.")
if_continue = input("ok that's it. I'm way too tired and I am off my shift now. Just pay the bill and go ok? (ok)")
print("let me just confirm this with you: ")
print("You ordered a", size, coffee_type, "with", flavor, "flavoring.")
print("The cost of your cup of coffee is $", coffee_cost, ".")
print("The total cost of your cup of coffee with a 15% tip is $", coffee_tip_cost, ".")

if baked_cost != 0:
    print("You also ordered", baked_amt, "banana breads with $20 per piece. Which is in total $", baked_cost)
elif mugs_cost != 0:
    print("You also ordered", mugs_amt, "mugs with $100 per piece. Which is in total $", mugs_cost)
print("your total cost is", coffee_tip_cost + mugs_cost + baked_cost)
if_extra_help = input("is there anything else I can do for you?")
if if_extra_help == "yes":
    print("oh I wish I can, but I don't want to :) Bye-bye")

```
